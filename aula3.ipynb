```python
# importando biblioteca pandas no Python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")

# Novo dataframe com colunas específicas
ca_df = combustiveis_df[['Estado - Sigla', 'Municipio', 'Revenda', 'Produto', 'Valor de Venda']]

# Somente Produto = gasolina
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
```

```python
# Usamos DataFrame.loc[] para criar filtros específicos.

etn_city = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'SAO PAULO')]
display(etn_city.sort_values(by='Valor de Venda'))
```

```python
# Busca média de preços da gasolina no bairro X da cidade de SÃO PAULO
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'LIBERDADE') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | 
                             (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), 
                            ['Valor de Venda']].mean())
```

```python
# groupby = agrupa produtos;
# mean = descobrir a média;
# round(x) = casas depois do ponto; 
# -----------------------------------------------------------------------------

# Média de valor de venda por combustível Brasil todo.
by_combustivel = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)

display(by_combustivel)
```
