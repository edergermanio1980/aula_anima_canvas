```python
# importando biblioteca pandas no Python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Adicionando uma nova coluna
combustiveis_df['Ativo'] = True

display(combustiveis_df.head(3))
```

```python
# Adiciona coluna 'Obs' que contenha "MELHOR CIDADE" quando a coluna `Municipio` = SAO PAULO
combustiveis_df['Obs'] = ["MELHOR CIDADE" if Municipio == 'SAO PAULO' else '' for Municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO', 'INDAIATUBA', 'CAMPINAS', 'SALTO']), ['Municipio', 'Obs']])
```

```python
# Se o valor de venda > 6,5 reais, retorna "Valor inacessível", caso contrário, "Valor acessível"
import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Inacessível', 'Acessível')
display(combustiveis_df[['Revenda', "Valor de Venda", 'Valor de Venda - Status']])
```

```python
# Calcula os Postos por habitante.

num_habitantes = pd.read_csv('ibge_num_habitantes_estimado.csv', sep=";")
num_habitantes.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes)
```

```python
# Realizar um MERGE dos dois dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
# limpar colunas nulas

merge_df.dropna(axis='columns', inplace=True)

print(merge_df.info())
```

```python
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Valor de Venda - Status']
merge_df.drop( labels=colunas, axis=1, inplace=True)

print(merge_df.info())
```

```python
# Remove linhas duplicadas
merge_df.drop_duplicates(inplace=True)

display(merge_df.head(100))
```

```python
# Agrupa e conta quantos postos tem por município

postos_by_municipios = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_by_municipios.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_by_municipios.rename({'Revenda': 'Número de Postos'}, inplace=True)
display(postos_by_municipios)
```
