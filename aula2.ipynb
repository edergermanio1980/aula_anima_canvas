```python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Exibe estastísticas

print(combustiveis_df.describe())
```

```python
# Busca coluna
print(combustiveis_df['Revenda'].tail(10))
```

```python
# Filtro no Dataframe
ca_df = combustiveis_df[['Estado - Sigla', 'Municipio', 'Revenda', 'Produto', 'Valor de Venda']]

display(ca_df.head(10))
```

```python
# Mostra linhas 4 a 9
display(ca_df.loc[4:9])
```

```python
# Filtra por tipo de combustível e mostra estatísticas
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']

display(gas_df.describe())

# Mostrar o local da gasolina mais barata
print("Gasolina com o maior valor")
display(gas_df.max())

# Mostrar o local da gasolina mais cara
print("\nGasolina com menor valor")
display(gas_df.min())
```
